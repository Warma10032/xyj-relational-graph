```python
# 根据人物列表作为关键词进行百度百科查询
import requests
from bs4 import BeautifulSoup
import urllib.parse
import csv
import pandas as pd
import time

input_csv_path = r'E:\tzy\code\py\pythonProject\person.csv'
output_csv_path = r"E:\tzy\code\py\pythonProject\attribute.csv"

# 打开 csv 文件
outcsv = open(output_csv_path, 'a', newline='', encoding='utf-8')

# 清空输出文件
with open(output_csv_path, 'w', newline='') as csvfile:
    # 使用空列表清空文件内容
    csvfile.write('')

# 设置表头，即你想要保存的属性，需与百度百科中的名称相同
fields = ['姓名','性别', '别名', '法宝','兵器', '神通','本相','技能','坐骑','住所']
csv_writer = csv.DictWriter(outcsv, fieldnames=fields)
csv_writer.writeheader()

# 读取 CSV 文件
df = pd.read_csv(input_csv_path)

# 将数据转换为列表
t_list = df.values.tolist()
person_list = []
for t in t_list:
    person_list.append(t[0])

url1 = 'https://baike.baidu.com/item/'
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36"
}

for key_word in person_list:
    time.sleep(0.3)
    html = requests.get(url1 + str(key_word), headers=headers)

    # 使用BeautifulSoup解析HTML
    soup = BeautifulSoup(html.content, 'html.parser')

    # 直接搜索具有 class='your-specific-class' 的 div 元素，此处class=?会随网页更新而变化，会失效。
    #（如何判断失效？在下面那个代码框中，print(dt_element)语句中进行debug，查看是否查找到）
    div_elements = soup.find_all('div', class_='itemWrapper_qVJDe') 

    # 存储结果的列表
    result_list = []
    item_dict={}
    
    item_dict['姓名']=key_word

    # 遍历每个 div 元素
    for div_element in div_elements:
        # 查找当前 div 元素下的 dt 和 dd 元素
        dt_element = div_element.find('dt')
        dd_element = div_element.find('dd')
    
        # 将 dt 和 dd 元素的文本内容组成一个属性字典
        if dt_element and dd_element:
            if str(dt_element.text.strip().replace('\xa0', '')) in fields:
                item_dict[str(dt_element.text.strip().replace('\xa0', ''))] = str(dd_element.text.strip())
        
    # 将一个人物的属性字典添加到结果列表
    result_list.append(item_dict)

    # 输出结果到控制台
    # 出现打印为空的原因。1.触发网站人机验证（通过下个代码块print(soup)语句debug）
    # 2. 网站class种类改变（通过下个代码块print(dt_element)语句debug），需要利用浏览器F12打开监视器查找
    print(result_list)

    # 将结果列表写入csv文件
    csv_writer.writerows(result_list)

# 关闭 CSV 文件
outcsv.close()

#下面代码可以没有，目的为了便于后续处理
# 读取 CSV 文件为 DataFrame
df = pd.read_csv(output_csv_path)

# 将缺失值替换为 NAN
df.fillna(value='NAN', inplace=True)

# 将 DataFrame 重新写入 CSV 文件
df.to_csv(output_csv_path, index=False, encoding='utf-8')

print("写入文件成功")
```
